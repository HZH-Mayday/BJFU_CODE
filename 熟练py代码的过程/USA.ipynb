```python
import pandas as pd
import numpy as np
```

```python
abbrevs = pd.read_csv("C:/Users/HuZheHui/Desktop/data-USstates-master/state-abbrevs.csv")
area = pd.read_csv("C:/Users/HuZheHui/Desktop/data-USstates-master/state-areas.csv")
population = pd.read_csv("C:/Users/HuZheHui/Desktop/data-USstates-master/state-population.csv")
abbrevs, area, population
```

```python
merge = pd.merge(abbrevs, population, how="outer", left_on="abbreviation", right_on="state/region")
merge.drop("abbreviation", axis=1, inplace=True)
merge.info()
# merge.loc[merge["state/region"].isnull(), :]
```

```python
merge.loc[merge["state"].isnull(), "state/region"].unique()
```

```python
merge.loc[merge["state/region"]=="USA", "state"] = "United States"
merge.loc[merge["state/region"]=="PR", "state"] = "Puerto Rico"
```

```python
final = pd.merge(merge, area, how="left", on="state")
final.head()
```

```python
final[final["area (sq. mi)"].isnull()]["state"].unique()
```

```python
final.dropna(inplace=True)
```

```python
data2010 = final.query("ages=='total' & year==2010")
data2010.head()
```

```python
data2010.set_index("state", inplace=True)
density = data2010["population"]/data2010["area (sq. mi)"]
density
```

```python
density.sort_values(ascending=False, inplace=True)
density.head()
```
