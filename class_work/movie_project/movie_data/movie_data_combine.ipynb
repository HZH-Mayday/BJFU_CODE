### 数据处理

```python
import pandas as pd
import numpy as np
df = pd.read_excel("./movie_info.xls", encoding="utf-8", index=False)
movie_df = df[["name", "date", "long", "rate", "comment_sum", "year", "direct_index", "actor_index", "country", "money"]]
movie_df
```

```python
# 填充缺失数据
movie_df.fillna({"long":int(movie_df["long"].mean()), "rate":movie_df["rate"].mean(), "comment_sum":movie_df["comment_sum"].mean()},
               inplace=True)
movie_df.info()
```

```python
movie_df["date"] = movie_df["date"].dt.strftime("%Y-%m-%d")
movie_df.info()
```

```python
# 档期分期
from datetime import datetime 
def Schedule_transform(date):
    day = date[5:]
    if "11-20"<=day<="12-31" and "01-01"<=day<="03-15":
        return "贺岁档"
    elif "04-28"<=day<="05-02":
        return "五一档"
    elif "06-01"<=day<="09-01":
        return "暑期档"
    elif "10-01"<=day<="10-07":
        return "十一档"
    else:
        return "普通"
movie_df["date"] = movie_df["date"].apply(Schedule_transform)
movie_df
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
country_label = LabelEncoder().fit_transform(movie_df["country"])
year_label = LabelEncoder().fit_transform(movie_df["year"])
year_onehot = OneHotEncoder().fit_transform(year_label[:, np.newaxis]).toarray()
date_label = LabelEncoder().fit_transform(movie_df["date"])
date_onehot = OneHotEncoder().fit_transform(date_label[:, np.newaxis]).toarray()
```

```python
# 整合数据
df_arr = movie_df.iloc[:, [2,3,4,6,7]].values
n = len(df_arr)
x = np.concatenate((df_arr, year_onehot, date_onehot, country_label.reshape(n, 1)), axis=1)
y = movie_df["money"].values
```

### 数据探索性分析

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
movie_df.describe()
```

```python
movie_df[["long","rate","comment_sum","year","direct_index","actor_index","money"]].applymap(lambda x: np.log(x)).corr()
```

```python
# np.hstack((x, y.reshape(400,1))).T
# pd.DataFrame(np.corrcoef())
xx = np.concatenate((StandardScaler().fit_transform((np.log(x[:, :5]+1))), x[:, 5:]) , axis=1 )
xy = np.hstack((xx, np.log(y.reshape(400,1)))).T
pd.DataFrame(np.corrcoef(xy))[21]
```

```python
model = LinearRegression().fit(xy.T[:, [0,1,2,3,4,15]], xy.T[:,-1])
model.score(xy.T[:, [0,1,2,3,4,15]], xy.T[:,-1])
```

```python
## 多项式
X_train, X_test, y_train, y_test = train_test_split(xy.T[:, :-1], xy.T[:, -1], test_size=0.33, random_state=42)

pol_tran = PolynomialFeatures(degree=3)
pol_model = pol_tran.fit(X_train)
pol_x_train = pol_model.transform(X_train)
pol_x_test = pol_model.transform(X_test)

model = LinearRegression().fit( pol_x_train, y_train)
model.score( pol_x_test, y_test )
```
