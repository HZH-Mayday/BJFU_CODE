```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
data_path = "C:/Users/Administrator/Desktop/draw_data.xls"
data = pd.read_excel(data_path,encoding="utf-8",sheetname=0).applymap(np.log)[::-1]
```

```python
data.index = [str(i) for i in data.index]

fig = plt.figure(0)
data["ln(RE)"].plot(c='black',linestyle='-', marker='o',linewidth=0.6,label="LNRE")
data["ln(UR)"].plot(c='black',linestyle='-',marker='v',linewidth=0.6,label="LNUR")
data["ln(RU)"].plot(c='black',linestyle='-',marker='*',linewidth=0.6,label="LNRU")
data["ln(GDP)"].plot(c='black',linestyle='-',marker='s',linewidth=0.6,label="LNGDP")
plt.legend(loc='center right', bbox_to_anchor=(0.75, -0.19),ncol=2)
```

```python
fig = plt.figure(0)
data["ln(RE)_1"].plot(c='black',linestyle='-', marker='o',linewidth=0.6,label="LNRE")
data["ln(UR)_1"].plot(c='black',linestyle='-',marker='v',linewidth=0.6,label="LNUR")
data["ln(RU)_1"].plot(c='black',linestyle='-',marker='*',linewidth=0.6,label="LNRU")
data["ln(GDP)"].plot(c='black',linestyle='-',marker='s',linewidth=0.6,label="LNGDP")
plt.legend(loc='center right', bbox_to_anchor=(0.75, -0.19),ncol=2)
```
