```python
#-*- coding: utf-8 -*-
%matplotlib inline
import numpy as np
from numpy import sin,cos,abs,sqrt
import matplotlib.pyplot as plt
```

```python
def love1(x):
    change = sqrt(1-x**2)+(x**2)**(1.0/3)
    return change

def love2(x):
    change =(x**2)**(1.0/3)-sqrt(1-x**2)
    return change
```

```python
love1(1)
love1(-1)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
x = np.linspace(-1,1,1000)
y = love1(x)
yy = love2(x)
ax.plot(x,y,'r',linewidth = 3)
ax.plot(x,yy,'r',label = u'小胡',linewidth = 3)
ax.plot(x/2,y/2,'#FF3030',linewidth = 2)
ax.plot(x/2,yy/2,'#FF3030',label=u'圈圈',linewidth = 2)
ax.grid()
ax.legend(loc=4,prop={'family':'SimHei','size':15})
fig.savefig('C:/Users/Administrator/Desktop/1.png',dip = 200)
ax.set_title(u"情人节快乐",fontproperties='SimHei')
```
