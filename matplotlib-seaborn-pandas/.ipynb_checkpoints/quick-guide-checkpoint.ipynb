```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
%matplotlib inline
```

```python
df = pd.read_excel("https://github.com/chris1610/pbpython/blob/master/data/sample-salesv3.xlsx?raw=true")
df.head()
```

#### 做一下聚合和整理，得到需要plot的数据

```python
top_10 = df.groupby("name")["ext price", "quantity"].agg({
    "ext price": "sum",
    "quantity": "count"
}).sort_values(
    by="ext price", ascending=False)[:10].reset_index()
top_10.rename(
    columns={
        "name": "Name",
        "ext price": "Sales",
        "quantity": "Purchases"
    },
    inplace=True)
top_10
```

#### 使用ggplot风格的画图风格

```python
plt.style.available
```

```python
plt.style.use("ggplot")
```

```python
fig, ax = plt.subplots()
top_10.plot(kind="barh", x = "Name", y = "Sales", ax=ax)
```

```python
fig, ax = plt.subplots()
top_10.plot(kind="barh", x = "Name", y = "Sales", ax=ax)
ax.set_xlim([-10000,140000])
ax.set_xlabel("Total Revenue")
ax.set_ylabel("Customer")
```

```python
fig, ax = plt.subplots()
top_10.plot(kind="barh", y="Sales", x="Name", ax=ax)
ax.set_xlim([-10000,140000])
ax.set(xlabel="Total Revenue", ylabel="Customer", title="2014 Revenue")
```

#### 隐藏标签legend

```python
fig, ax = plt.subplots()
top_10.plot(kind="barh", y="Sales", x="Name", ax=ax)
ax.set_xlim([-10000,140000])
ax.set(xlabel="Total Revenue", ylabel="Customer", title="2014 Revenue")
ax.legend().set_visible(False)
```

#### 去掉网格化， 并且额外添加一条辅助线， 并在辅助线上添加文字

```python
fig, ax = plt.subplots(figsize=(5,6))
top_10.plot(kind="barh", y="Sales", x="Name", ax=ax)
avg = top_10["Sales"].mean()
ax.set_xlim([-10000,140000])
ax.set(xlabel="Total Revenue", ylabel="Customer", title="2014 Revenue")
ax.axvline(x=avg, color="blue", label="Average", linestyle="--", linewidth=1)  # 添加辅助线

for cut in [3,5,8]:
    ax.text(x=avg+1000, y=cut, s="New Customer",color="green", bbox=dict(facecolor='red', alpha=0.3))

ax.legend().set_visible(False) 
ax.grid(False) # 去掉网格
```

#### 调整x轴坐标的刻度

```python
def currency(x, pos):
    '''x表示value，pos表示调整的位置'''
    if x>1000000:
        return "${:.1f}M".format(x*1e-6)
    return "${:.0f}K".format(x*1e-3)

fig, ax = plt.subplots()
top_10.plot(kind="barh", y="Sales", x="Name", ax=ax)
ax.set_xlim([-10000,140000])
ax.set(xlabel="Total Revenue", ylabel="Customer", title="2014 Revenue")
formatter = FuncFormatter(currency)
ax.xaxis.set_major_formatter(formatter)
ax.legend().set_visible(False)
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=1, ncols=2, sharey=True, figsize=(10,4))
fig.suptitle("2014 Sales Analysis", fontsize=14, fontweight="bold")
top_10.plot(kind="barh", y="Sales", x="Name", ax=ax0)
ax0.set(title="Revenue", xlabel="Total revenue", ylabel="Customer")
formatter = FuncFormatter(currency)
ax0.xaxis.set_major_formatter(formatter)
ax0.axvline(x=top_10["Sales"].mean(), color="blue", linestyle="--", linewidth=1)
ax0.legend().set_visible(False)

top_10.plot(kind="barh", y="Purchases", x="Name", ax=ax1)
ax1.axvline(x=top_10["Purchases"].mean(), color="blue", linestyle="--", linewidth=1)
ax1.set(title="Units", xlabel="Total units", ylabel="")
ax1.legend().set_visible(False)
```
