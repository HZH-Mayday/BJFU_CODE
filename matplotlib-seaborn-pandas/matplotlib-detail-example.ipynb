```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
%matplotlib inline
```

### 导入数据

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data
target = iris.target
data = np.concatenate((features, target.reshape(len(target),1)), axis=1)
```

```python
df = pd.DataFrame(data, columns=['sl','sw','pl','pw',"target"])
df.head()
```

### 普通线图和气泡图

```python
x = np.linspace(1,10,100)
y1 = np.sin(x)
y2 = np.cos(x)
fig, ax = plt.subplots()
ax.plot(x,y1,c="b",linestyle="-",linewidth=1, label="sin(x)")
ax.plot(x,y2,c="g",linestyle="-.",linewidth=1, label="cos(x)")
ax.set(title="math fomular", xlim=[1,10], ylim=[-1,1], xlabel="x", ylabel="y")
ax.legend(loc=1, frameon=False)     # 去掉legend的边框
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)
# 当color是数字
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar() # 显示颜色条
```

### DataFrame的调入法

```python
fig, ax = plt.subplots()
# cmap参数可选plasma, inferno, magma, viridis
df.plot.scatter(x="sl", y="sw", c=target, s=df["pl"]*50, alpha=0.5, cmap="magma", ax=ax)
```

```python
fig, ax = plt.subplots()
df.plot.hist(bins=20, histtype='stepfilled', alpha=0.3, density=True, ax=ax)
```

### 图例的设置方法

```python
x = np.linspace(0, 10, 1000)
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), '-b', label='Sine')
ax.plot(x, np.cos(x), '--r', label='Cosine')
ax.axis("equal")      # 设置轴的属性
ax.legend(loc=1, frameon=False, ncol=2)       # 去掉图例的边框，并且把图例分成两列
```

```python
x = np.linspace(0, 10, 1000)
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), '-b', label='Sine')
ax.plot(x, np.cos(x), '--r', label='Cosine')
ax.axis("equal")      # 设置轴的属性
# fancybox表示边框是否圆润， framealpha表示透明度， shadow表示是否有阴影， borderpad表示间隔
ax.legend(fancybox=True, framealpha=1, shadow=True, borderpad=1, ncol=2)
```

#### 不同的尺寸大小的legend

```python
fig, ax = plt.subplots()
for area in [100, 300, 500]:
    ax.scatter([], [], c='k', alpha=0.3, s=area, label=str(area) + ' km$^2$')
# scatterpoint表示legend的点个数， labelspacing表示每行的距离
ax.legend(scatterpoints=1, frameon=False, labelspacing=1, title='City Area')
```

### 文字的位置

```python
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10])
# transData(默认)表示按照数据选区位置
# transAxes按照坐标轴位置
# transFigure按照整个图片维度为单位
ax.text(1, 5, ". Data: (1, 5)", transform=ax.transData)
ax.text(0.5, 0.1, ". Axes: (0.5, 0.1)", transform=ax.transAxes)
ax.text(0.2, 0.2, ". Figure: (0.2, 0.2)", transform=fig.transFigure)
```

### 画箭头

```python
x = np.linspace(0, 10, 1000)
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), '-b', label='Sine')
ax.plot(x, np.cos(x), '--r', label='Cosine')
ax.axis("equal")      # 设置轴的属性
ax.legend(loc=1, frameon=False, ncol=2) 
ax.annotate("max", xy=(np.pi/2+4, 1), xytext=(6,-2), size=15,
            arrowprops=dict(arrowstyle="fancy",fc="0.4", ec="none",connectionstyle="angle3,angleA=0,angleB=-90"),
           bbox=dict(boxstyle="round", alpha=0.1))
```
