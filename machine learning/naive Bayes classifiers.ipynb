### [朴素贝叶斯的三种形式](https://www.letiantian.me/2014-10-12-three-models-of-naive-nayes/)
* 高斯朴素贝叶斯算法

```python
from sklearn.datasets import make_blobs
x, y = make_blobs(n_samples=100, n_features=2, centers=2, random_state=2, cluster_std=1.5)
plt.scatter(x[:,0], x[:,1], c=y, s=50, cmap="RdBu")
```

```python
# 建立模型
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x, y)
```

```python
rng = np.random.RandomState(0)
xnew = [-6, -14] + [14, 18] * rng.rand(2000, 2)
ynew = model.predict(xnew)
plt.scatter(x[:,0], x[:,1], c=y, cmap="RdBu", s=50)
lim = plt.axis()
plt.scatter(xnew[:,0], xnew[:,1], c=ynew, cmap="RdBu", s=20, alpha=0.1)
plt.axis(lim)
```

```python
# 提取每一个数据在类中的概率
y_pro = model.predict_proba(xnew)
y_pro[-8:].round(2)
```
